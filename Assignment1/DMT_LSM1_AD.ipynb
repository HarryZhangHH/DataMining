```python
# create template for join
data_template = df.drop_duplicates()
data_template = data_template['time']
data_template = data_template.drop_duplicates()
data_template = pd.DataFrame(data_template)
data_template
```

```python
# collect all variables for one subject of one day IN ONE ROW
for key in d.keys():
# process subjects from 14.01 to 14.33
    data_all_per_subject = dict(tuple(d[key].groupby('variable'))) 
    d_tem = data_template
    for var in dict_var.keys():
        # take each variable and create a new column for a subject 
        try:
            data_per_var = data_all_per_subject[var][["time","value"]]
            data_per_var = data_per_var.groupby('time').mean()
            data_per_var = data_per_var.rename(columns = {"value" : var})
            d_tem = d_tem.join(data_per_var, on = 'time')
        except KeyError as e:
        # if one subject do not have a set of data (i.e., AS14.02 do not record finance), create an empty df and join!
            data_per_var = pd.DataFrame(columns = [var])
            d_tem = d_tem.join(data_per_var)
    d_tem.to_csv('out_' + key + '.csv')
```

```python
# merge all data of each subject together
df_all = pd.read_csv('out_AS14.01.csv')
for key in d.keys():
    if key != 'AS14.01':
        df_sub = pd.read_csv('out_' + key + '.csv')
        df_all = pd.concat([df_all, df_sub], axis=0, sort=False)
df_all.to_csv('out_allsub.csv')
```
